```python
customer_reviews = ["sam was a great help to me in the store",
"the cashier was very rude to me, I think her name was eleanor", "amazing work from sadeen!",
"sarah was able to help me find the items I needed quickly",
"lucy is such a great addition to the team",
"great service from sara she found me what i wanted"]
```

```python
sarah_review=[]
```

```python
pattern_to_find=r"sarah?"
```

```python
for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        sarah_review.append(string)
```

```python
print(sarah_review)
```

```python
#^ - mean staring with this sentence
a_review=[]
pattern_to_find=r"^a"
for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        a_review.append(string)

print(a_review)
```

```python
#$ - mean ending with this sentence
y_review=[]
pattern_to_find_Y_ending=r"y$"
for string in customer_reviews:
    if(re.search(pattern_to_find_Y_ending,string)):
        y_review.append(string)

print(y_review)
```

Or this or this word on this sentence

```python
needwant_reviws = []
```

```python
pattern_to_find = r"(need|want)ed"
for string in customer_reviews:
    if(re.search(pattern_to_find,string)):
        needwant_reviws.append(string)

print(needwant_reviws)
```

Removing Punctuation

```python
no_punct_reviews = []
```

```python
# []:a set of characters we want
# ^ = "not"
# \w = word characters
# \s =  whitespace
#sub() = replacing text
pattern_to_find=r"[^\w\s]"
```

```python
for string in customer_reviews:
    no_punct_string = re.sub(pattern_to_find," ",string)
    no_punct_reviews.append(no_punct_string)
```

```python
print(no_punct_reviews)
```

PreProcessing

Stemming reduce the number of unique words
Reduce the complexcity

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens=['connecting','connect','connectivity','connects']
```

```python
for t in connect_tokens:
    print(t,":",ps.stem(t))
```

```python
likes_token = ['Likes','Better','Worse']
for t in likes_token:
    print(t,":",ps.stem(t))
```

N-grams help us analyze the relationship between neighboring words

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
token = [
    'the', 'rise', 'to', 'significant', 'advancements', 'natural', 'artificial', 'intelligence', 'has',
    'led', 'to', 'advancements', 'in', 'language', 'processing', 'computer', 'vision', 'and', 'other',
    'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated',
    'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought',
    'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of',
    'deep', 'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful',
    'capable', 'of', 'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from',
    'it', 'in', 'ways', 'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai',
    'is', 'increasingly', 'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from',
    'healthcare', 'to', 'finance', 'to', 'transportation', 'and', 'its', 'impact', 'is', 'only',
    'set', 'to', 'grow', 'in', 'the', 'years', 'to', 'come'
]

print(token)
```

```python
unigrams =(pd.Series(nltk.ngrams(token,1)).value_counts())
print(unigrams[:10])
```

```python
unigrams[:10].sort_values().plot.barh(
    color="lightblue",
    width=0.75,
    figsize=(12, 7)
)
plt.title("10 Most Frequently Occurring Unigrams")
plt.xlabel("Count")
plt.ylabel("Unigrams")
plt.show()
```

```python
bigrams =(pd.Series(nltk.ngrams(token,2)).value_counts())
print(bigrams[:10])
```

```python
trigrams =(pd.Series(nltk.ngrams(token,3)).value_counts())
print(trigrams[:10])
```
