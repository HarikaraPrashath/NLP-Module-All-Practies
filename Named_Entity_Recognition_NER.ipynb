Named Entity Recognition (NER)

Detects names of people, organizations, locations, etc.

ðŸ‘‰ Example: "Elon Musk founded SpaceX" â†’ Elon Musk (PERSON), SpaceX (ORG)

```python
from html.entities import html5
from xml.dom.minidom import Entity

import  spacy
from spacy import tokenizer
from spacy import displacy
from IPython.display import HTML,display
import re
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
google_text = (
    "google was founded on september 4, 1998, by computer scientists larry page and sergey brin while they were phd "
    "students at stanford university in california. together they own about 14% of its publicly listed shares and "
    "control 56% of its stockholder voting power through super-voting stock. the company went public via an initial "
    "public offering (ipo) in 2004. in 2015, google was reorganized as a wholly owned subsidiary of alphabet inc. "
    "google is alphabet's largest subsidiary and is a holding company for alphabet's internet properties and interests. "
    "sundar pichai was appointed ceo of google on october 24, 2015, replacing larry page, who became the ceo of alphabet. "
    "on december 3, 2019, pichai also became the ceo of alphabet."
)
print(google_text)
```

```python
spacy_doc = nlp(google_text)
```

```python
'''
.label = a numeric ID of the Entity
.label_ = a human-readable label
'''
for word in spacy_doc.ents:
    print(word.text,word.label_)
```

```python
html = displacy.render(spacy_doc, style="ent",jupyter=False)
display(HTML(html))
```

```python
google_text_clean = re.sub(r'[^\w\s]','',google_text).lower()
print(google_text_clean)
```

```python
spacy_doc_clean = nlp(google_text_clean)
```

```python
for word in spacy_doc_clean.ents:
    print(word.text,word.label_)
```
